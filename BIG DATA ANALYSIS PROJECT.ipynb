```python
# Introduction to Big Data
# Tool used: PySpark
    
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("BigDataAnalysis") \
    .getOrCreate()
```

```python
# Load the Dataset
df = spark.read.csv("netflix_titles.csv", header=True, inferSchema=True)
df.show(5)
 
# Basic Data Exploration
df.printSchema()
df.describe().show()
df.count()
```

```python
# Data Preprocessing
# Data Cleaning 
# Remove null values
df = df.dropna()

# Drop duplicates
df = df.dropDuplicates()
```

```python
# Big Data Analysis
# Analysis (grouping, aggregation, filtering)
# Group By Analysis
df.groupBy("category").count().show()

# Aggregation
from pyspark.sql.functions import avg

df.groupBy("category").agg(avg("sales")).show()

#Filtering
df.filter(df["sales"] > 5000).show()
```

```python
# Scalability Feature
print(df.rdd.getNumPartitions())
```

```python
# Optional Visualization
import matplotlib.pyplot as plt

pdf = df.toPandas()

pdf['category'].value_counts().plot(kind='bar')
plt.show()
```
